```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pe(pos, d = 100):
    encoding = np.zeros(d)
    for i in range(d):
        if i % 2 == 0:
            encoding[i] = np.sin(pos/10000**((2*i)/d))
        else:
            encoding[i] = np.cos(pos/10000**((2*i)/d))
    return encoding

for i in range(50):
    plt.plot(pe(i))
```

```python
from numpy import dot
from numpy.linalg import norm

def cos_sim(a, b):
    return dot(a, b)/(norm(a)*norm(b))
```

```python
for i in range(10):
    print(i, cos_sim(pe(10),pe(10-i)), np.dot(pe(10),pe(10-i)))
```

```python
for i in range(10):
    print(i, cos_sim(pe(10),pe(10+i)), np.dot(pe(10),pe(10+i)))
```

```python
for i in range(10):
    print(i, cos_sim(pe(10000),pe(10000-i)), np.dot(pe(10000),pe(10000-i)))
```
